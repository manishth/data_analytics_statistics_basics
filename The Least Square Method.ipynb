### Tips for service example:

```python
total_bill = [34, 108, 64, 88, 99, 51]
tip_amount = [5, 17, 11, 8, 14, 5]
```

tip_amount is DEPENDENT on the total_bill amount. So, tip_amount is DEPENDENT and total_bill is INDEPENDENT value.

LEAST SQUARES CRITERION:

formula: 

##### <b>min &Sigma; (y - y&macr;) ^ 2 </b>

here:
<i>y</i> = observed value of dependent variable (tip amount)<br/>
<i>y&macr;</i> = estimated (predicted) value of the dependent variable (predicted tip amount)<br/>
<i>min</i> = minimum value<br/>

<i>So, the goal is to minimize the sum of the squared differences between the observed value (tip_amount) and the predicted value (predicted tip-amount) that is provided by the regression line. Sum of the squared residuals.</i>

```python
import numpy as np
```

```python
total_bill_mean = np.mean(total_bill)
tip_amount_mean = np.mean(tip_amount)
```

##### CENTROID:

    centroid = (total_bill_mean, tip_amount_mean) 
    
The best fit regression line will/must pass through the centroid

##### y = mx + c

here:

<i>m</i>: is the slope<br/>
<i>c</i>: is the intercept

```python
# Let's calculate the slope and intercept

from scipy.stats import linregress
linregress(total_bill, tip_amount)
```

From above we see that:

Slope (m) = 0.1462 <br/>
Intercept (c) = -0.8202

##### What does this mean?

For every `$`1 the bill amount(x) increases, we would expect the tip amount to increase by `$`0.1462 (slope m) or about 15 cents.

If the bill amount(x) is zero, then the expected or predicted tip amount `$`-0.8202 (intercept c) or negative 82 cents! Does this make sense? NO. Thuse the intercept may or may not make sense in the 'real world'.
